# Step 6: Testing Hypotesis

**Hypothesis to Test**

*"The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."*

***

**Null Hypothesis (H0):** The average ride duration from the Loop to O’Hare does not differ on rainy Saturdays compared to other Saturdays.

**Alternative Hypothesis (H1):** The average ride duration from the Loop to O’Hare is different on rainy Saturdays.

***

Since we’re testing for a difference (rather than specifically *“greater”* or *“less”*), this is a **two-sided** test.

````python
#First, let's import the DataFrame and have a quick check

df_rides = pd.read_csv("/datasets/project_sql_result_07.csv")

df_rides.info()
df_rides.head()
````

### Data Cleaning and Feature Extraction

**Spliting the Data into Two Groups**

- Group 1: Durations on rainy Saturdays
- Group 2: Durations on non-rainy Saturdays

### Choosing an Statistical Test

````python
from scipy.stats import ttest_ind

df_rides = pd.read_csv('/datasets/project_sql_result_07.csv')

# Convert 'start_ts' to datetime
df_rides['start_ts'] = pd.to_datetime(df_rides['start_ts'])

# Extract day of the week (0 = Monday, 5 = Saturday, 6 = Sunday)
df_rides['day_of_week'] = df_rides['start_ts'].dt.dayofweek

# Identify Saturdays
df_rides['is_saturday'] = df_rides['day_of_week'] == 5

# Filter rides for Saturdays with "Good" and "Bad" weather conditions
saturday_rainy = df_rides[(df_rides['is_saturday']) & (df_rides['weather_conditions'] == 'Bad')]['duration_seconds']
saturday_non_rainy = df_rides[(df_rides['is_saturday']) & (df_rides['weather_conditions'] == 'Good')]['duration_seconds']

# Perform the t-test (Welch's t-test assuming unequal variances)
t_stat, p_value = ttest_ind(saturday_rainy, saturday_non_rainy, equal_var=False, nan_policy='omit')

# Define significance level
alpha = 0.05

# Output results
{
    "p-value": p_value,
    "t-statistic": t_stat,
    "Decision": "Reject H0 (significant difference)" if p_value < alpha else "Fail to reject H0 (no significant difference)"
}
````

# Conclusion

- Since we rejected H0, we conclude that ride duration changes on bad-weather Saturdays.
- The positive t-statistic (7.18) suggests that rides **tend** to be longer in bad weather compared to good weather.

***
Since p-value < 0.05, we reject H0 and conclude that ride duration is significantly different between good and bad weather Saturdays. The positive t-statistic (7.18) suggests that rides take longer on bad-weather Saturdays.

This result aligns with expectations, as *traffic congestion, slower driving speeds*, and *longer pickup times* could contribute to increased ride durations in bad weather.
